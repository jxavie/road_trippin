```python
# import dependencies
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# import traffic volume data
file_path = 'va_roads/raw/VDOT_Traffic_Volume.csv'
traffic_volume = pd.read_csv(file_path, encoding='utf8', low_memory=False)
traffic_volume.head()
```

```python
traffic_volume['DATA_DATE'] = traffic_volume['DATA_DATE'].astype('datetime64')

traffic_volume['Year'] = pd.DatetimeIndex(traffic_volume['DATA_DATE']).year
traffic_volume.head()
```

```python
# import pavement condition data
file_path = 'va_roads/raw/VDOT_Pavement_Conditions_2019_MOD.csv'
pavement_condition = pd.read_csv(file_path, encoding='utf8', low_memory=False)
pavement_condition.head()
```

```python
# pavement_condition['Year'] = pavement_condition['Year'].astype('int64')
pavement_condition['Year'] = pd.to_datetime(pavement_condition['Year'])
pavement_condition['Year'].head()
```

```python
combined = pavement_condition.merge(traffic_volume, left_on=['Year', 'ROUTE_NAME'], right_on=['Year', 'ROUTE_NAME'])
combined.head()
```

```python
# import dependencies
import pandas as pd
```

```python
# assign federal dataset path to variable
csv_path = "expenditures/federal_v_state/1_cleaned/federal_total_CL-OM_2017dollars.csv"

# load csv
fed_breakdown = pd.read_csv(csv_path)

# transpose table
fed_breakdown = fed_breakdown.rename(columns={"Unnamed: 0":"Spending"})
fed_breakdown = fed_breakdown.set_index("Spending")
del fed_breakdown.index.name
fed_breakdown = fed_breakdown.transpose()
fed_breakdown.head()
```

```python
# assign state dataset path to variable
csv_path = "expenditures/federal_v_state/1_cleaned/state-local_total_CL-OM_2017dollars.csv"

# load csv
state_breakdown = pd.read_csv(csv_path)
state_breakdown.head()

# transpose table
state_breakdown = state_breakdown.rename(columns={"Unnamed: 0":"Spending"})
state_breakdown = state_breakdown.set_index("Spending")
del state_breakdown.index.name
state_breakdown = state_breakdown.transpose()
state_breakdown.head()
```

```python
# assign total (federal + state and local) dataset path to variable
csv_path = "expenditures/federal_v_state/1_cleaned/total_CL-OM_2017dollars.csv"

# load csv
total_breakdown = pd.read_csv(csv_path)

# transpose table
total_breakdown = total_breakdown.rename(columns={"Unnamed: 0":"Spending"})
total_breakdown = total_breakdown.set_index("Spending")
del total_breakdown.index.name
total_breakdown = total_breakdown.transpose()
total_breakdown = total_breakdown.dropna(axis=1, how='all')
total_breakdown = total_breakdown.drop("Unnamed: 63")
total_breakdown.head()
```

```python
# assign gdp percentage dataset path to variable
csv_path = "expenditures/federal_v_state/1_cleaned/total_gdp_2018dollars.csv"

# load csv
gdp_percent = pd.read_csv(csv_path)

# transpose table
gdp_percent = gdp_percent.set_index("Year")
del gdp_percent.index.name
gdp_percent = gdp_percent.dropna(axis=1, how='all')

# conversion factor (2018 dollars to 2017 dollars)
factor = 361/369.8

# create columns with 2017 dollar values
gdp_percent["Federal (Billions of 2017 Dollars)"] = gdp_percent["Federal (Billions of 2018 Dollars)"] * factor
gdp_percent["State and Local (Billions of 2017 Dollars)"] = gdp_percent["State and Local (Billions of 2018 Dollars)"] * factor
gdp_percent = gdp_percent[["Federal (Billions of 2017 Dollars)",
                           "State and Local (Billions of 2017 Dollars)",
                           "Federal (% GDP)",
                           "State and Local (% GDP)"]]
gdp_percent.head()
```

```python
fed_breakdown.to_csv("expenditures/federal_v_state/fed_expenditures_2017dollars.csv")
state_breakdown.to_csv("expenditures/federal_v_state/state_expenditures_totals_2017dollars.csv")
total_breakdown.to_csv("expenditures/federal_v_state/total-fed-state_expenditures_2017dollars.csv")
gdp_percent.to_csv("expenditures/federal_v_state/expenditures_percent-gdp_2017dollars.csv")
```
